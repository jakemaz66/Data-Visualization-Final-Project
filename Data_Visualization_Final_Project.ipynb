Changing 'streams' column from object to integer type

```python
#Dropping problematic 'streams' row, there is one error with a long incorrect value
print(df[df['streams'].str.len() > 15])

df = df[df['streams'].str.len() < 15]
```

```python
#Changing the type of 'streams' column to integer
df['streams'] = df['streams'].astype(int).copy()
```

Which year yields the highest number of streams?

```python
#Grouping the data frame by the released year
year_groups = df.groupby('released_year')['streams'].mean()
```

```python
#Histogram of raw counts of songs from each year
sns.histplot(data=df, x='released_year', bins=20)
plt.title('Raw counts of Songs in Released in each Year')
```

```python
#Top 10 most streamed years
top_10 = df.groupby('released_year')['streams'].mean()

#Resetting the index and making it most to least order
top_10 = top_10.sort_values(ascending=False).reset_index()

#Indexing the top 10 values
top_10 = top_10[:10]
```

```python
top_10
```

```python
#Barchart to show the average stream values for 10 highest years
sns.barplot(data=top_10, x='released_year', y='streams')
plt.title('Average Streams for top 10 Released Years')
plt.xlabel('Released Year')
plt.ylabel('Streams')
```

High Speechiness seems to correlate with lower streams

```python
#Finding the p-value for correlation with speechiness and streams using scipy library
correlation, p_value = scipy.stats.pearsonr(df['speechiness_%'], df['streams'])

#Printing values
print(f"Pearson Correlation: {correlation}")
print(f"P-Value: {p_value}")

#Setting significance threshold
alpha = 0.05

if p_value < alpha:
    print("There is a statistically significant correlation.")
else:
    print("There is no statistically significant correlation.")
```

Which month is most popular?

```python
df['released_month']
```

```python
#Function to change the integer of months into actual names, then applying to the released_month column
def change_month_name(month):
    month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    return month_names[month - 1]

df['released_month'] = df['released_month'].apply(change_month_name)
```

```python
df['released_month']
```

```python
#Making plotly barchart to show the streams by month
month = df.groupby('released_month')['streams'].mean().reset_index()

df = df.merge(month, on='released_month', suffixes=('', '_mean'))
#Renaming the columns
df.rename(columns={'streams_mean': 'average_streams'}, inplace=True)

#Creating a plotly barchart
fig = px.bar(df, x='released_month', y='average_streams', title='Monthly Average Streams')

#Updating Axis Titles
fig.update_layout(
    xaxis_title='Month',
    yaxis_title='Average Streams',
    yaxis_tickformat=',d',
)

fig.show()
```
